```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set(style= 'darkgrid')
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data= tips)
```

```python
tips['smoker'].value_counts()
```

```python
sns.relplot(x = 'total_bill', y= 'tip', data= tips, hue= 'smoker', style= 'time')
```

```python
sns.relplot(x = 'total_bill', y = 'tip', hue='size', data= tips)
```

```python
sns.relplot(x = 'total_bill', y = 'tip', hue= 'size', data=tips, palette= 'ch:r = -0.5, l = 0.75')
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data=tips, size='size')
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data=tips, size='size', sizes = (15, 200))
```

```python
from numpy.random import randn
```

```python
df = pd.DataFrame(dict(time = np.arange(500), value = randn(500).cumsum()))
df.head()
```

```python
sns.relplot(x = 'time', y = 'value', kind= 'line', data=df)
```

```python
df = pd.DataFrame(randn(500, 2).cumsum(axis=0), columns= ['time', 'value'])
df.head()
```

```python
sns.relplot(x = 'time', y = 'value', data= df, kind='line')
```

```python
sns.relplot(x = 'time', y = 'value', data = df, kind = 'line', sort = False)
```

```python
fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
sns.relplot(x = 'timepoint', y = 'signal', kind = 'line', data = fmri)
```

```python
sns.relplot(x = 'timepoint', y = 'signal', kind= 'line', data = fmri, ci = False)
```

```python
sns.relplot(x = 'timepoint', y = 'signal', kind = 'line', data = fmri, ci = 'sd')
```

```python
sns.relplot(x = 'timepoint', y = 'signal', kind = 'line', data = fmri, estimator = None)
```

```python
sns.relplot(x = 'timepoint', y = 'signal', hue = 'event', kind = 'line', data = fmri)
```

```python
sns.relplot(x = 'timepoint', y = 'signal', hue = 'region', style = 'event', kind = 'line', data = fmri)
```

```python
sns.relplot(x = 'timepoint', y = 'signal', hue = 'region', style = 'event', kind = 'line', data = fmri, markers = True, dashes = False)
```

```python
sns.relplot(x = 'timepoint', y = 'signal', hue = 'event', style = 'event', kind = 'line', data = fmri)
```

```python
sns.relplot(x = 'timepoint', y = 'signal', hue = 'region', units = 'subject', kind = 'line', estimator = None, data = fmri.query("event == 'stim'"))
```

```python
dots = sns.load_dataset('dots').query("align == 'dots'")
dots.head()
```

```python
sns.relplot(x = 'time', y = 'firing_rate', hue = 'coherence', data = dots, kind = 'line', style = 'choice')
```

```python
pallete = sns.cubehelix_palette(light= 0.5, n_colors=6)
sns.relplot(x = 'time', y = 'firing_rate', hue = 'coherence', data = dots, kind = 'line', style = 'choice', palette=pallete)
```

```python
sns.relplot(x = 'time', y = 'firing_rate', size = 'coherence', hue = 'coherence', style = 'choice', kind = 'line', data = dots, sizes=(1, 5))
```
