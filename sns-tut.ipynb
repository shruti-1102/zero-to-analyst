```python
dots = sns.load_dataset('dots').query("align == 'dots'")
dots.head()
```

```python
sns.relplot(x = 'time', y = 'firing_rate', hue = 'coherence', data = dots, kind = 'line', style = 'choice')
```

```python
pallete = sns.cubehelix_palette(light= 0.5, n_colors=6)
sns.relplot(x = 'time', y = 'firing_rate', hue = 'coherence', data = dots, kind = 'line', style = 'choice', palette=pallete)
```
